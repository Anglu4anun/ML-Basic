## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
import re

def convert_case(string, force_format=None):
    # определение формата
    if force_format is None:
        # если первая буква строчная то snake_case
        default_format = 'snake' if string[0].islower() else 'pascal'
    else:
        default_format = force_format.lower()

    # Snake_case to PascalCase
    if default_format == 'snake':
        words = string.split('_')
        pascal_case = ''.join(word.capitalize() for word in words)
        return pascal_case

    # PascalCase to snake_case
    elif default_format == 'pascal':
        snake_case = re.sub(r'([A-Z])', r'_ \\1', string).strip().replace(' ', '').lower()
        return snake_case.lstrip('_')

    raise ValueError("nеправильный формат")

print(convert_case("this_is_snake_case"))          # Output: ThisIsSnakeCase
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
import math
def is_prime(n):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
users_data = {}

while True:
    print("\\nВведите данные пользователя:")
    first_name = input("Имя: ").strip()
    last_name = input("Фамилия: ").strip()
    age_input = input("Возраст: ")
    id_input = input("ID: ")

    if not first_name or not last_name or not age_input or not id_input:
        break
    # верхний регистр
    first_name = first_name.capitalize()
    last_name = last_name.capitalize()
    # возраст
    try:
        age = int(age_input)
        if age < 18 or age > 60:
            raise ValueError("Возраст должен быть от 18 до 60.")
    except ValueError as e:
        print(f"Ошибка: {e}")
        continue
    # ID
    try:
        user_id = int(id_input)
        formatted_id = f"{user_id:08}"  # Доводим до 8 цифр нулевыми символами слева
        if formatted_id in users_data:
            raise ValueError("Данный ID уже занят!")
    except ValueError as e:
        print(f"Ошибка: {e}")
        continue
    # добавляем
    users_data[formatted_id] = (first_name, last_name, age)
print("\\nДанные введены:\\n")
for key, value in users_data.items():
    print(f"ID: {key}, Имя: {value[0]}, Фамилия: {value[1]}, Возраст: {value[2]}")
```
